```python
categorical=train[['Product_id','Customer_name','Loyalty_customer','Product_Category']]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label=LabelEncoder()
cate_data=categorical.apply(label.fit_transform)
```

```python
cate_data.head()
```

```python
train.drop(['Product_id','Customer_name','Loyalty_customer','Product_Category','instock_date'],axis=1,inplace=True)
```

```python
frames=[train,cate_data]
new_data=pd.concat(frames,axis=1)
```

```python
new_data.head()
```

```python
new_data=new_data.apply(pd.to_numeric,errors='coerce')
```

```python
new_data.head()
```

```python
new_data.dtypes
```

```python
new_data.isna().sum()
```

```python
new_data=new_data.dropna()
```

```python
new_data.isna().sum()
```

```python
X=new_data.drop(['Selling_Price'],axis=1)
y=new_data['Selling_Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.22,random_state=42)
```

```python
from sklearn import linear_model
```

```python
model=linear_model.LinearRegression()
model.fit(X_train,y_train)
```

```python
print('The test_score is',model.score(X_test,y_test))
print('The train_score is',model.score(X_train,y_train))
```

```python
pred=model.predict(X_test)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse=mean_squared_error(y_test,pred)
```

```python
rmse=np.sqrt(mse)
```

```python
print("The rmse score is ",rmse)
```

```python
print("The R2 score is",r2_score(y_test,pred))
```

```python
print('The test_score is',model.score(X_test,y_test))
print('The train_score is',model.score(X_train,y_train))
print("The root_mean_squred score is ",rmse)
print("The R2 score is",r2_score(y_test,pred))
```
